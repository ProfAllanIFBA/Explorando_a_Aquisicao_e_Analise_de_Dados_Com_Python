#### Exemplo 1: Gráfico de Funções

```python
# Bibliotecas e Módulos
import matplotlib.pyplot as plt
import numpy as np

# Dados
x = np.arange(0,2*np.pi,0.01)
y1 = np.sin(x) # Função seno
y2 = np.cos(x) # Função cosseno


# Exibição de grade
plt.grid()

# Adicionando linhas de referência (eixos x e y)
plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

# O método plot() define os eixos do gráfico
plt.plot(x,y1,'-',color = 'blue', label = 'sen(x)') # Troque o marcador x por :, d, o, ., .-
plt.plot(x,y2,'-',color = 'red', label = 'cos(x)') # Troque o marcador x por :, d, o, ., .-

# Título do Gráfico e Rótulos dos Eixos
plt.title('Função Trigonométricas Básicas')
plt.xlabel('x')
plt.ylabel('y')

# Exibindo o Gráfico
plt.show()
```

#### Exemplo 2: Gráficos Estatísticos

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset tips
tips = sns.load_dataset("tips")

# Configuração do estilo
sns.set(style="whitegrid")

# Criar um grid de plots com 2 colunas
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Scatter plot
sns.scatterplot(x="total_bill", y="tip", hue="day", size="size", palette="viridis", data=tips, ax=axs[0])
axs[0].set_title("Scatter Plot: Total Bill vs Tip")

# Box plot
sns.boxplot(x="day", y="total_bill", hue="smoker", palette="coolwarm", data=tips, ax=axs[1])
axs[1].set_title("Box Plot: Total Bill by Day")

# Violin plot
sns.violinplot(x="time", y="total_bill", hue="sex", split=True, palette="muted", data=tips, ax=axs[2])
axs[2].set_title("Violin Plot: Total Bill by Time and Sex")

# Ajustar a disposição dos plots
plt.tight_layout()
plt.show()
```

#### Exemplo 3: Gráficos 3D

```python
import numpy as np
import plotly.graph_objs as go
from sympy import symbols, lambdify, parse_expr

def plot_3d_surf_cart(expressions, x_min, x_max, passo_x, y_min, y_max, passo_y):
    x, y = symbols('x y')
    f = [lambdify((x, y), parse_expr(expr), "numpy") for expr in expressions]
    x_vals, y_vals = np.arange(x_min, x_max, passo_x), np.arange(y_min, y_max, passo_y)
    x_mesh, y_mesh = np.meshgrid(x_vals, y_vals)
    z_mesh = [func(x_mesh, y_mesh) for func in f]

    traces = [go.Surface(x=x_mesh, y=y_mesh, z=z, colorscale='hot', showscale=False) for z in z_mesh]

    layout = go.Layout(
        title='Gráfico 3D',
        scene=dict(xaxis=dict(title='X Label'), yaxis=dict(title='Y Label'), zaxis=dict(title='Z Label')),
        width=800, height=800
    )

    fig = go.Figure(data=traces, layout=layout)

    def atualizar_grafico(angle_y, angle_z):
        angle_y, angle_z = np.radians(angle_y), np.radians(angle_z)
        cos_y, sin_y, cos_z, sin_z = np.cos(angle_y), np.sin(angle_y), np.cos(angle_z), np.sin(angle_z)

        rot_y = np.array([[cos_y, 0, sin_y], [0, 1, 0], [-sin_y, 0, cos_y]])
        rot_z = np.array([[cos_z, -sin_z, 0], [sin_z, cos_z, 0], [0, 0, 1]])
        rot_matrix = rot_y @ rot_z

        for i, z in enumerate(z_mesh):
            rotated = np.dot(rot_matrix, np.vstack((x_mesh.flatten(), y_mesh.flatten(), z.flatten()))).reshape(3, x_mesh.size)
            fig.data[i].update(x=rotated[0], y=rotated[1], z=rotated[2])

    fig.show()
```

```python
 plot_3d_surf_cart(['x**2+y','x**2+y**2'], -5, 5, 0.1, -5, 5, 0.1) 
```

Operadores são símbolos especiais usados em programação para realizar operações aritméticas, lógicas e relacionais em valores e variáveis. Em Python, há vários tipos de operadores que podem ser usados para manipular valores e realizar cálculos. #### <div style="margin-top: -20px;">

<font color='blue'><span style="font-size: 16px;"><strong>4.3.1 Operadores Aritméticos</strong></span></font>

<hr style="border: 2px solid blue;">
</div>

Alguns exemplos de **operadores aritméticos** em Python incluem:

$+$ (**adição**): adiciona dois valores juntos.

$-$ (**subtração**): subtrai um valor do outro.

$*$ (**multiplicação**): multiplica dois valores juntos.

$/$ (**divisão**): divide um valor pelo outro. 

$**$ (**potência**): retorna a potência de um valor.

```python
# Exemplos
a,b,c,d,e = 12,3,7,-4,2 #Atribuição em linha
```

```python
# Operações básicas +,-,*,/
soma = a + b
sub = b - d
prod = a*d
div = b/c
pot = a**e
```

```python
# Impressão das variáveis armazenadas
print(soma)
print(sub)
print(prod)
print(div)
print(pot)
```

#### <div style="margin-top: -20px;">
<hr style="border: 2px solid blue;">

<font color='blue'><span style="font-size: 16px;"><strong>4.3.2 Operadores Relacionais</strong></span></font>

<hr style="border: 2px solid blue;">
</div>

Alguns exemplos de **operadores relacionais** em Python incluem:

$==$ (**igual a**): verifica se dois valores/variáveis são iguais.

$!=$ (**diferente de**): verifica se dois valores/variáveis são diferentes.

$<$ (**menor que**): verifica se um valor é menor que outro. (variação $<=$ **menor ou igual a**)

$>$ (**maior que**): verifica se um valor é maior que outro. (variação $>=$ **maior ou igual a**)

```python
# Exemplos
print(2 == 1+1)
print('Allan' != 'allan')
print(4.1 < 4)
print(5 < 5)
print(5 <= 5)
print({1,2,3} == {1,3,2})
print((1,2,3) == (1,3,2))
```

#### <div style="margin-top: -20px;">
<hr style="border: 2px solid blue;">

<font color='blue'><span style="font-size: 16px;"><strong>4.3.3 Operadores Lógicos</strong></span></font>

<hr style="border: 2px solid blue;">
</div>

Alguns exemplos de **operadores lógicos** em Python incluem:

**and** (**e**): retorna True se ambos os valores são verdadeiros.

**or** (**ou**): retorna True se pelo menos um dos valores é verdadeiro.

**not** (**não**): inverte o valor booleano de uma expressão.

Seguem as tabelas lógicas da conjunção negação **not**, **and** e disjunção **or**:<br><br>
$$
\begin{array}{|c|c|c|c|c|}
\hline
p	        & 	 not \,\, p \\
\hline
 True  		& 	 False   	\\
 False  	& 	 True  	\\
\hline
\end{array} \,\,\,
\begin{array}{|c|c|c|c|c|}
\hline
p	        & 	 q   	    & p \,\, and \,\, q   \\
\hline
 True  		& 	 True   	& True   	\\
 True 		& 	 False   	& False	    \\
 False  	& 	 True   	& False  	\\
 False  	& 	 False   	& False    	\\
\hline
\end{array} \,\,\,

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carrega a imagem
imagem = cv2.imread('img_apostila\pessoas.jpeg')  # Substitua 'pessoas.jpeg' pelo caminho correto da sua imagem

# Define os intervalos de cores para segmentação (exemplo: segmentar a cor vermelha)
limite_inferior = np.array([0, 0, 100])
limite_superior = np.array([100, 100, 255])

# Cria uma máscara para segmentar a cor dentro do intervalo especificado
mascara = cv2.inRange(imagem, limite_inferior, limite_superior)

# Aplica a máscara na imagem original
imagem_segmentada = cv2.bitwise_and(imagem, imagem, mask=mascara)

# Converte a imagem de BGR para RGB para exibição correta no Matplotlib
imagem_segmentada_rgb = cv2.cvtColor(imagem_segmentada, cv2.COLOR_BGR2RGB)

# Mostra a imagem segmentada usando Matplotlib
plt.imshow(imagem_segmentada_rgb)
plt.title('Imagem Segmentada')
plt.axis('off')  # Oculta os eixos
plt.show()
```

```python
# Imports necessários
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import numpy as np
from PIL import Image
import pickle
import os
import matplotlib.pyplot as plt

# Função para carregar imagens das pastas
def load_images_from_folder(base_folder):
    data = []
    labels = []
    label_dict = {'label_3': 0, 'label_5': 1}  # Gato = 0, Cachorro = 1
    for label_folder, label in label_dict.items():
        folder_path = os.path.join(base_folder, label_folder)
        for filename in os.listdir(folder_path):
            if filename.endswith('.png'):
                img_path = os.path.join(folder_path, filename)
                img = Image.open(img_path)
                img = np.array(img)
                data.append(img)
                labels.append(label)
    return np.array(data), np.array(labels)

# Carregar dados e criar arquivo batch
base_folder = 'imagens_treino'
data, labels = load_images_from_folder(base_folder)
output_file = 'batch_de_treino'
with open(output_file, 'wb') as file:
    pickle.dump({'data': data, 'labels': labels}, file, protocol=pickle.HIGHEST_PROTOCOL)

# Função para carregar dados personalizados a partir de um arquivo batch
def load_custom_data(batch_file):
    with open(batch_file, 'rb') as file:
        batch = pickle.load(file)
        imagens = np.array(batch['data'])
        labels = np.array(batch['labels'])
    return imagens, labels

# Caminho para o arquivo batch
batch_file = 'batch_de_treino'

# Carregar os dados do arquivo batch
imagens, labels = load_custom_data(batch_file)

# Embaralhar os dados
indices = np.arange(len(imagens))
np.random.shuffle(indices)
imagens = imagens[indices]
labels = labels[indices]

# Dividir os dados em conjuntos de treino (80%) e teste (20%)
num_treino = int(len(imagens) * 0.8)
imagens_treino, imagens_teste = imagens[:num_treino], imagens[num_treino:]
labels_treino, labels_teste = labels[:num_treino], labels[num_treino:]

# Normalizar os dados
imagens_treino = imagens_treino / 255.0
imagens_teste = imagens_teste / 255.0

# Construção do modelo ajustado
nosso_modelo = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(2, activation='softmax')  # Ajustar para 2 classes
])

# Compilar o modelo
nosso_modelo.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Treinar o modelo
history = nosso_modelo.fit(imagens_treino, labels_treino, epochs=20, validation_data=(imagens_teste, labels_teste))

# Avaliar o modelo
erro_teste, acc_teste = nosso_modelo.evaluate(imagens_teste, labels_teste, verbose=2)
print('\nAcurácia com Dados de Teste:', acc_teste)
```

```python
# Função de previsão ajustada
def prever(img_png):
    # Carregar a nova imagem
    nova_imagem = Image.open(f"imagens_verificacao/{img_png}")
    
    # Verificar se a imagem tem um canal alfa (RGBA) e convertê-la para RGB se necessário
    if nova_imagem.mode == 'RGBA':
        nova_imagem = nova_imagem.convert('RGB')
    
    # Redimensionar a imagem para 32x32 pixels
    nova_imagem = nova_imagem.resize((32, 32))
    
    # Converter a imagem para um array NumPy e normalizar os valores dos pixels
    nova_imagem_array = np.expand_dims(np.array(nova_imagem) / 255.0, axis=0)
    
    # Fazer a previsão usando o modelo treinado
    previsoes = nosso_modelo.predict(nova_imagem_array)
    
    # Obter a classe prevista (índice da classe com maior probabilidade)
    classe_prevista = np.argmax(previsoes)
    
    # Definir os nomes das classes
    classes = ['Gato', 'Cachorro']
    
    # Exibir a imagem com o título da classe prevista
    plt.figure(figsize=(2,2))
    plt.imshow(nova_imagem)
    plt.title(f"Classificada como: {classes[classe_prevista]}")
    plt.axis('off')
    plt.show()
```

Mas onde entra Python nesta história? O código a seguir troca qualquer quantidade de espaços por apenas um em um texto copiado de um pdf. Em seguida ele abre o translate automaticamente mostrando a tradução correta.

```python
import webbrowser
import urllib.parse
import re

def traduzir_texto_para_portugues(texto_original):
    def corrigir_espacos(texto):
        # Substitui múltiplos espaços por um único espaço
        return re.sub(r'\s+', ' ', texto)

    # Codifica o texto para URL após corrigir os espaços
    texto_url = urllib.parse.quote_plus(corrigir_espacos(texto_original))
    
    # Cria a URL diretamente para o Google Tradutor com auto-detecção de idioma de origem e destino como português
    url = f"https://translate.google.com/?sl=auto&tl=pt&text={texto_url}&op=translate"
    
    # Abre a URL no navegador padrão
    webbrowser.open(url)
```

```python
def newton_raphson(f, df, x0, tol=1e-5, max_iter=100):
    """
    Método de Newton-Raphson para encontrar a raiz de uma função.
    
    :param f: Função cuja raiz queremos encontrar.
    :param df: Derivada da função f.
    :param x0: Aproximação inicial.
    :param tol: Tolerância para o critério de parada.
    :param max_iter: Número máximo de iterações.
    :return: Aproximação da raiz.
    """
    xn = x0
    for n in range(max_iter):
        fxn = f(xn)
        dfxn = df(xn)
        if abs(fxn) < tol:
            print(f"Convergência atingida após {n+1} iterações.")
            return xn
        if dfxn == 0:
            print("Derivada zero. Nenhuma solução encontrada.")
            return None
        xn = xn - fxn / dfxn
        print(f"Iteração {n+1}: x = {xn}, f(x) = {fxn}")
    print("Número máximo de iterações alcançado. Nenhuma solução encontrada.")
    return xn

# Função e sua derivada
f = lambda x: x**2 - 2
df = lambda x: 2*x

# Aproximação inicial
x0 = 1

# Chamando o método de Newton-Raphson
raiz = newton_raphson(f, df, x0)

print(f"A raiz encontrada é: {raiz}")
```

Particularmente, eu buscaria, por exemplo, aplicar o método de Newton a equações não lineares básicas, aquelas que não se necessita deles. Estas equações são úteis pois pode-se comparar o resultado exato com o resultado fornecido. Por exemplo, 

* $x^{2} - 5x + 6 = 0$ (raízes $x_{1} = 2$ e $x_{2} = 3$);
* $e^{x} - 2 = 0$ (raiz $x = ln(2) = 0.6931471805599453$);
* $sen(x) - e^{x} +1 = 0$ (Uma raíz é $x = 0$).

O último dos exemplos acima, embora não se tenha um método específico usou o fato de $sen(0) = 0$ e $e^{0} = 1$ para gerar uma equação que tenha raíz igual a $0$.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 2

def df(x):
    return 2 * x

def newton_raphson_visual(f, df, x0, tol=1e-5, max_iter=5):
    x_vals = np.linspace(0, 2, 400)
    y_vals = f(x_vals)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_vals, label='$f(x) = x^2 - 2$')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

    xn = x0
    for n in range(max_iter):
        fxn = f(xn)
        dfxn = df(xn)
        if abs(fxn) < tol:
            break
        if dfxn == 0:
            break
        # Calculate the tangent line
        tangent_line = fxn + dfxn * (x_vals - xn)
        plt.plot(x_vals, tangent_line, linestyle='--', color='gray')
        plt.scatter([xn], [fxn], color='red')  # Current approximation
        # Newton-Raphson step
        xn_new = xn - fxn / dfxn
        plt.plot([xn, xn_new], [fxn, 0], 'ro-')  # Line to the x-axis
        xn = xn_new
        plt.text(xn, f(xn), f'$x_{{{n+1}}}$', fontsize=12, color='blue')
    
    plt.scatter([xn], [0], color='blue', zorder=5)  # Final approximation
    plt.title('Método de Newton-Raphson')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()

# Aproximação inicial
x0 = 1

# Visualização do método de Newton-Raphson
newton_raphson_visual(f, df, x0)
```
